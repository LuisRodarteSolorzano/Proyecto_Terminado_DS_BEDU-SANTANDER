#### Resumen
El principal objetivo de este análisis es investigar la potencial relación entre la pandemia del Covid-19 y un mercado financiero particular como lo es el de las criptomonedas, específicamente en la pionera y más importante, el Bitcoin. Después de declarada la pandemia en marzo del 2020, el Bitcoin tuvo el mínimo para todo el 2020 pero al final del año había incrementado su valor 500\%. El mercado de criptomonedas se caracteriza por fluctuaciones inciertas y tendencias poco claras. Aquí análizamos la relación entre los retornos y volatilidad del bitcoin con indicadores del impacto social de la pandemia del Covid-19 (casos, muertes, índice de sentimiento, índice global del miedo, cobertura de los medios de comunicación, y noticias falsas).

#### Motivación e investigación preliminar 
El año 2020 fue caracterizado por un evento sin precedentes en el ámbito de la salud, la pandemia del Covid-19. Trás su declaración como pandemia mundial el 11 de marzo 2020, por el _World Health Organization (WHO)_, se tomaron medidas de aislamiento social para evitar la propagación generando un importante shock en la economía global y mercados financieros particulares, esto debido a la incertidumbre entre el público que los lleva a retirar posiciones en inversión para buscar liquidez y dejando en posición vulnerable los precios en los mercados financieros involucrados.

El mercado de criptomonedas no ha sido la excepción ya que, con la mencionada declaración de pandemia al virus del Covid-19 el 11 de Marzo 2020, la capitalización de mercado del Bitcoin toco un pico mínimo, para todo el 2020, con un respectivo precio de 4,970 USD. Sin embargo, para finales del mismo mes de Marzo ya había recuperado su precio de antes de la caída y fue superándola hasta alcanzar casi el valor de 30,000 USD al final del año, lo que representa un incremento de 500\%, en un año caracterizado por la crisis, dejando un efecto ininteligible, e incluso se podría pensar que positivo, del Covid-19 en el mercado. Para el año 2021 ya se podía apreciar una recuperación, a nivel global, de los casos y muertes reportados por Covid-19 generando de a poco optimismo en la sociedad, sin embargo las fluctuaciones en el mercado de criptomonedas resulto incierto y sin alguna tendencia clara durante el periodo.

Dada la magnitud del impacto del Covid-19 en el estilo de vida de la sociedad en general y de la tendencia a resguardarse en casa, ha surgido una importante atención en los medios ya que la población está alerta ante cualquier aviso o actualización sobre el virus. Según información reportada por "Ravenpack" el 70% de las noticias globales estaban hablando sobre el Covid-19 ya sea informando o generando pánico ante las cifras de muertes reportadas. Ante esta situación, además de analizar el Covid-19 mediante casos y muertes reportadas a nivel global, es posible utilizar también indicadores de medios sociales con el Covid-19 como tópico común.

En recientes investigaciones, como las de Umar Zaghum, _"A time–frequency analysis of the impact of the Covid-19 induced panic on the volatility of currency and cryptocurrency markets"_ y _"The impact of COVID-19-related media coverage on the return and volatility connectedness of cryptocurrencies and fiat currencies"_ en 2021, entre otras, se ha mostrado que durante el período de máximo auge del Covid-19 se presentó un importante impacto de los indicadores de pánico y cobertura de medios en el mercado de criptomonedas y el mercado de divisas tradicional (caso específico de Europa y China), este último impacto es distinto al presentado para las criptomonedas por lo que sugiere el autor que podría presentarse, en las criptomonedas, una posibilidad de cobertura ante el riesgo para protección del patrimonio de inversionistas.

## **Colección de Datos**

1. La serie del BTC se obtuvo de [yahoo finanzas](https://es-us.finanzas.yahoo.com/) a través de la librería yfinance, extrayendo la información diaria del 01 de Enero 2020 al 31 de Diciembre 2021. Trabajaremos con la serie del BTC convertida a retornos que, usualmente en la literatura son calculados de la siguiente forma:

\begin{equation}
r_{btc}=\log (\frac{btc_{it}}{btc_{it-1}}) \times 100
\end{equation}

2. Se construyó el índice GFI siguiendo a los autores (Salisu 2020, Rubbaniy 2021) con las bases de datos de muertes y casos reportados, extraído de la _United Nations Office for the Coordination of Humanitarian Affairs (OCHA)_ en el sitio https://data.humdata.org/dataset/novel-coronavirus-2019-ncov-cases, con datos desde el 22 de Enero 2020 al 31 de Diciembre 2021. Mide el nivel de miedo en la población con base en la severidad del Covid-19, la construcción se realizó como se detalla a continuación:

\begin{equation}
ICR_{t} = \left(\frac{\sum_{i}^{N}{c_{i,t}}}{\sum_{i}^{N}{\left(c_{i,t}+c_{i,t-5}\right)}}\right)\times 100
\end{equation}


\begin{equation}
IMR_{t} = \left(\frac{\sum_{i}^{N}{d_{i,t}}}{\sum_{i}^{N}{\left(d_{i,t}+d_{i,t-5}\right)}}\right) \times 100
\end{equation}

En donde:\
$i=1,2,\cdots, N$; &nbsp; &nbsp;$t=1,2,\cdots, T$; &nbsp; &nbsp; $ICR: $ índice de casos reportados; $c_{i,t}: $ casos reportados en todos los $i$ países en el tiempo (284 en la muestra) $t$; $IMR: $ índice de muertes reportadas; $d_{i,t}: $ muertes reportadas en todos los $i$ países en el tiempo $t$.\
Y, finalmente, el índice global de miedo será el promedio de los anteriores índices.

\begin{equation}
GFI_{t} = \left(\frac{1}{2}\left(ICR_{t}+IMR_{t}\right)\right)
\end{equation}

3. Los indicadores “panic_index”, “media_hype”, “sentiment index”, “fake news”, “media_coverage_index” e "infodemic_index" se utilizarán de la base de datos "Ravenpack", la cual es utilizada por un gran número de autores en papers de investigación, el cuál está basado en más de 20 mil fuentes de noticias a nivel global y construyen con técnicas de análisis de sentimiento y machine learning indicadores relacionados al Covid-19, los cuales pueden ser consultados en el sitio https://coronavirus.ravenpack.com/ y detallaremos su definición a continuación:

    i. Panic index (PI): Mide el nivel de conversación en las noticias que hace referencia al pánico o histeria y coronavirus. Los valores están en porcentaje de 0 a 100. \
    ii. Media Hype (MH): Mide el porcentaje de noticias que hablan sobre el tema de coronavirus. Los valores están en porcentaje de 0 a 100.\
    iii. Sentiment index (SI): Mide el nivel de sentimiento en las entidades que son mencionadas, en las noticias, junto con el coronavirus. Los valores están entre -100 y 100 en donde 100 representa el sentimiento más positivo, -100 el más negativo y 0 neutral.\
    iv. **Media Coverage Index (MCI)**: Calcula el porcentaje de todas las fuentes de noticias que cubren el tema del coronavirus. Los valores están en porcentaje entre 0 y 100.\
    iv. **Infodemic index**: Calcula el porcentaje de las entidades (lugares, companias, etc) que de alguna manera se asocian al Covid-19. Los valores están en porcentaje entre 0 y 100 donde un valor de 60 significa que el 60% de las entidades cubiertas por los medios están asociadas al Covid-19.

Ahora vamos a definir una función que utilizaremos para elaborar los heatmaps.

```python
def heatmap(dataframe,titulo):
    # Creamos heatmap con seaborn
  fig, ax = plt.subplots(figsize = (10,15))

  fig.suptitle(titulo, size = 25, y = 0.92)
  heatmap_corr2 = sns.heatmap(data = dataframe,
                           ax = ax,
                           annot = True, # anotar los valores dentro de los recuadros
                           annot_kws = {'size':13}, # Cambiar tamaño de letra de
                                                    # dentro de recuadros
                           cmap = "PiYG",
                           center = 0, 
                           vmin = -1,
                           vmax = 1,
                           square = True, # Que sean cuadrados y no rectangulos
                           cbar = True,
                           cbar_kws = {'orientation': 'horizontal',
                                       'pad': 0.17},
                           linewidths = 5, # Grosor de las lineas que separan 
                                           # los recuadros
                           linecolor = 'white' # por defecto es 'white'
                           )

    # Quitar lineas de ticks del heatmap
  ax.tick_params(axis='both', length=0, labelsize=12)

    # Extraer del heatmap el colorbar
  cbar_corr = ax.collections[0].colorbar
    # Agregar titulo al colorbar
  cbar_corr.set_label('Correlación de Pearson', size=18)
    # Cambiar las ticks del colorbar
  cbar_corr.ax.tick_params(axis='both', length=0, labelsize=10, pad=5)
```

```python
gfi = pd.Series(periodo_critico2['gfi'])
pi = pd.Series(periodo_critico2['panic_index'])
mhi = pd.Series(periodo_critico2['media_hype_index'])
si = pd.Series(periodo_critico2['sentiment_index'])
mci = pd.Series(periodo_critico2['media_coverage_index'])
rbtc = pd.Series(periodo_critico2['Retornos_BTC'])
rbtcbin = pd.Series(periodo_critico2['RetonosBTC_bin'])
```

```python
lr = LinearRegression()
```

Definiremos una función para realizar los pasos de la regresion con cada par:

```python
def regresion(x, y, id_modelo,valor_comprobar):
  #MODELO ENTRENADO
  lr.fit(x.to_frame(),y)
  y_predict_id_modelo = lr.predict(x.to_frame())
  #GRAFICOS
  sns.scatterplot(x=x,y=y).set(title='Originales vs Predichos');
  sns.scatterplot(x=x,y=y_predict_id_modelo);
  #ECUACION
  intercept = lr.intercept_
  coefficient = lr.coef_[0]
  print(f'y = {coefficient} * x + {intercept}')
  #COMPROBACION
  print(f'Valor de x: {x[valor_comprobar-1]}')
  print(f'Valor predicho por nuestro modelo: {y_predict_id_modelo[valor_comprobar-1]}')
  print(f'Valor cuantificado directamente usando la funcion: {coefficient * x[valor_comprobar-1] + intercept}')
  #R2
  print(f'R2: {lr.score(x.to_frame(), y)}')
```

```python
#Modelo 1 y = GFI X = pi
regresion(pi, gfi, 'm1',10)
```

Con estos resultados podemos decir que nuestra variable de gfi(Indice global de miedo) es bien explicada por la variable de pánico ya que de acuerdo con el coeficiente del modelo podemos decir que por cada incremento unitario en el nivel de pánico el índice global de miedo aumentará en .010 unidades en promedio, dato congruente desde la definición de estos indicadores.\
Es notable observar que con el R2 podemos confirmar que nuestro modelo puede explicar un 10% de la variación de nuestra variable dependiente gfi, por lo que no podríamos quedarnos con este modelo tal como esta. 

```python
#Modelo 2 y=gfi X=media hype
regresion(mhi, gfi, 'm2', 10)
```

El resultado es muy similar que en el modelo 1, sin embargo en este caso es incluso menor el valor de R2 por lo que seguimos prefiriendo el primer modelo.

```python
#Modelo 3 y=gfi X=sentiment index
regresion(si, gfi, 'm3', 10)
```

Este caso, obtuvimos valores similares para los coeficientes que el modelo 1 y 2, sin embargo en este se obtiene un valor de R2 mejor.

```python
#Modelo 4 y=gfi X=media coverage
regresion(mci,gfi,'m4',10)
```

En este modelo el valor del coeficiente se mantiene similar a los otros modelos pero perdemos poder de predicción con el $R^2$ bajo.

```python
#Modelo 5 y=retornos btc X=media coverage
regresion(mci,rbtc,'m5',10)
```

En este modelo 5 el coeficiente nos indica que ante un incremento unitario en el indicador de cobertura de medios provocará un incremento de 0.03 unidades en los retornos del Bitcoin, en promedio.

```python
#Definiremos la función de bootstrap
def bstrap(var, n, r):
  serie = var
  media = []
  #Bootstrap
  for i in range(r):
    muestra = serie.sample(n=n, replace=True)
    media.append(muestra.mean())
  serie_medias = pd.Series(media)
  #Valores de intervalo de confianza
  limite_inferior = serie_medias.quantile(0.025)
  limite_superior = serie_medias.quantile(0.975)
  #Estadisticos del muestreo
  print(f'Error estandar: {serie_medias.std()}')
  print(f'Asimetria: {serie_medias.skew()}')
  print(f'Curtosis: {serie_medias.kurtosis()}')
  #Intervalo de confianza
  print(f'Intervalo de 95% confianza de la media: {limite_inferior} < {serie.mean()} < {limite_superior}')
  #Grafica de distribución con lineas de intervalo
  sns.distplot(serie_medias, kde=False, norm_hist=False)
  plt.axvline(limite_inferior)
  plt.axvline(limite_superior);
```
